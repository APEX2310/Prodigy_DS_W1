```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the data from the CSV file
file_path = r"C:\Users\saket\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_1897128\API_SP.POP.TOTL_DS2_en_csv_v2_1897128.csv"
data = pd.read_csv(file_path, skiprows=4)  # skipping the first 4 rows which usually contain metadata
```

```python
print(data.head())
```

## Creation of Histogram ##

```python
import matplotlib.pyplot as plt

# Select the data for the year 2020
year = '2020'
population_data = data[year].dropna()  # dropna() to remove any missing values

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_data, bins=30, edgecolor='black')
plt.title('Histogram of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()

```

## creation of Bar Graph ##


```python
# Assume the latest year is 2020 (adjust as needed based on your data)
latest_year = '2020'

# Select the data for the latest year and sort by population
top_10_countries = data[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(10)

# Create the bar graph
plt.figure(figsize=(12, 8))
plt.bar(top_10_countries['Country Name'], top_10_countries[latest_year], color='skyblue')
plt.title(f'Top 10 Most Populous Countries in {latest_year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
